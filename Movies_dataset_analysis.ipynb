```python
# Drop irrelevant columns
df.drop(['vote_count', 'status', 'adult', 'backdrop_path', 'homepage', 'imdb_id', 'original_language', 'original_title', 'overview', 'poster_path', 'tagline', 'production_countries'], axis='columns', inplace=True)
```

```python
# Change release_date to datetype
df['release_date'] = pd.to_datetime(df['release_date'])
```

### Process columns , to split the text with seperator as ',' to get the first value and integrate into the df

```python
genre = []
for row in df['genres'].str.split(','):
    genre.append(row[0])
```

```python
print(len(genre))
print(df['genres'].shape[0])
```

```python
df['genres'] = genre
```

```python
df.head()
```

```python
prod_companies = []
for row in df['production_companies'].str.split(','):
    prod_companies.append(row[0])
```

```python
print(len(prod_companies))
print(df['production_companies'].shape[0])
```

```python
df['production_companies'] = prod_companies
```

```python
df.head()
```

```python
langs = []
for row in df['spoken_languages'].str.split(','):
    langs.append(row[0])
```

```python
print(len(langs))
print(df['spoken_languages'].shape[0])
```

```python
df['spoken_languages'] = langs
```

```python
df.head()
```

```python
df.rename(columns={'vote_average':'rating', 'production_companies': 'production_house',
                 'spoken_languages':'language'}, inplace=True)
```

```python
df.drop(['popularity'], axis='columns', inplace=True)
```

```python
df.head()
df.to_csv('tmdb_cleaned.csv')
```
